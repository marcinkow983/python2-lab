# Broadcasting — zastosowania biznesowe

```python
# Ceny 4 produktów w 3 sklepach
ceny = np.array([[29.99, 31.99, 28.99],    # Produkt A
                  [49.99, 52.99, 47.99],    # Produkt B
                  [14.99, 15.99, 13.99],    # Produkt C
                  [99.99, 104.99, 94.99]])  # Produkt D

sklepy = ['Centrum', 'Galeria', 'Online']
produkty = ['A', 'B', 'C', 'D']
print(f"Kształt macierzy: {ceny.shape}")  # (4, 3)
```

Zadanie 1a: Stwórz nowy array z cenami podniesionymi o 5% (inflacja). Wyświetl wynik.

```python
print(ceny * 1.05)
```

Zadanie 1b: Stwórz array z cenami po rabacie. Każdy sklep daje inny rabat: Centrum 10%, Galeria 5%, Online 15%.

```python
rabaty = np.array([0.10, 0.05, 0.15])
mnozniki = 1 - rabaty
ceny_po_rabacie = ceny * mnozniki

print("Macierz cen po uwzględnieniu rabatów dla poszczególnych sklepów:")
print(ceny_po_rabacie)
```

Zadanie 1c: Oblicz koszt zakupu każdego produktu. Każdy produkt ma inną marżę: A=30%, B=40%, C=25%, D=50%.

```python
marza = np.array([[0.30], [0.40], [0.25], [0.50]])
koszt_zakupu = ceny / (1 + marza)
print("Macierz kosztów zakupu dla każdego produktu (wiersze A, B, C, D):")
print(koszt_zakupu)
```

Zadanie 1d: Wyświetl najtańszy sklep dla każdego produktu.

```python
sklepy = np.array(['Centrum', 'Galeria', 'Online'])
indeksy_najtanszych = ceny.argmin(axis=1)
najtansze_sklepy = sklepy[indeksy_najtanszych]
print("Indeksy najtańszych sklepów dla produktów A, B, C, D:")
print(indeksy_najtanszych)

print("\nNazwy najtańszych sklepów:")
print(najtansze_sklepy)
```

 # Reshape i łączenie danych 

Zadanie 2a — Reshape

```python
# Dane miesięczne z jednego roku (flat)
dane_flat = np.array([120, 135, 98, 142, 167, 189, 201, 178, 156, 143, 198, 221])

# Przekształć na kwartały (4 kwartały × 3 miesiące)
kwartaly = dane_flat.reshape(4, 3)
print(f"Kwartały:\n{kwartaly}")
print(f"Suma per kwartał: {kwartaly.sum(axis=1)}")
print(f"Najlepszy kwartał: Q{kwartaly.sum(axis=1).argmax() + 1}")
```

Zadanie 2b — Stacking

```python
# Dane z 3 oddziałów (osobne tablice)
warszawa = np.array([150, 160, 170, 180])
krakow = np.array([120, 130, 140, 150])
wroclaw = np.array([90, 100, 110, 120])

# Połącz w macierz (3 oddziały × 4 kwartały)
wszystkie = np.vstack([warszawa, krakow, wroclaw])
print(f"Wszystkie:\n{wszystkie}")
print(f"Suma per oddział: {wszystkie.sum(axis=1)}")
print(f"Suma per kwartał: {wszystkie.sum(axis=0)}")
```

Zadanie 2c — Praktyczne

```python
sprzedaz_flat = np.array([100, 200, 150, 300, 250, 180, 120, 220, 170, 310, 260, 190])
sprzedaz_macierz = sprzedaz_flat.reshape(6, 2)
srednia_per_produkt = sprzedaz_macierz.mean(axis=1)
zmiana = sprzedaz_macierz[:, 1] - sprzedaz_macierz[:, 0]

print("Macierz sprzedaży (Produkt x Miesiąc):")
print(sprzedaz_macierz)

print("\nŚrednia sprzedaż dla każdego produktu:")
print(srednia_per_produkt)

print("\nZmiana sprzedaży (Miesiąc 2 - Miesiąc 1):")
print(zmiana)
```

Zadanie 2d — Flatten i concatenate

```python
sprzedaz_wektor = sprzedaz_macierz.flatten()
nowy_miesiac = np.array([235])
sprzedaz_rozszerzona = np.concatenate([sprzedaz_wektor, nowy_miesiac])

print("Wektor po spłaszczeniu (12 miesięcy):")
print(sprzedaz_wektor)

print("\nWektor po dodaniu 13. miesiąca:")
print(sprzedaz_rozszerzona)
```

# Ćwiczenie 3: Analiza danych finansowych

Dane

```python
np.random.seed(2026)

# 8 pracowników, dane roczne
pracownicy = ['Anna', 'Jan', 'Ewa', 'Marek', 'Kasia', 'Piotr', 'Zofia', 'Tomek']
pensje = np.array([5500, 7200, 4800, 9100, 6300, 4200, 8500, 5800])
staz = np.array([3, 8, 2, 15, 5, 1, 12, 4])
oceny = np.array([4.2, 3.8, 4.5, 4.0, 4.7, 3.5, 4.3, 3.9])

# Sprzedaż miesięczna 4 produktów przez 6 miesięcy
produkty_nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Akcesorium']
sprzedaz = np.random.randint(20, 200, size=(4, 6))
ceny_produktow = np.array([3500, 1800, 2500, 150])
```

Zadanie 3a: Analiza wynagrodzeń

```python
sr_pensje = pensje.mean()
mediana = np.median(pensje)
std_pensje = pensje.std()

print(sr_pensje)
print(mediana)
print(std_pensje)


nazwy_arr = np.array(pracownicy)
powyzej_sredniej = nazwy_arr[pensje > sr_pensje]
print("Osoby zarabiające powyżej średniej: ", powyzej_sredniej)

ponizej_mediany = nazwy_arr[pensje < mediana]
print("Osoby zarabiające poniżej mediany: " ,ponizej_mediany)

rozstep = pensje.max() - pensje.min()
iqr = np.percentile(pensje, 75) - np.percentile(pensje, 25)

print("Rozstęp (Range): " ,rozstep)
print("Rozstęp międzykwartylowy (IQR): " ,iqr)
```

Zadanie 3b: System premii (np.where)

```python
premie = np.where(oceny >= 4.5, 2000, np.where(oceny >= 4.0, 1000, 0))

print("Przyznane premie:")
print(premie)

nowe_wynagrodzenia = pensje + premie

print("\nNowe wynagrodzenia (suma):")
print(nowe_wynagrodzenia)

pracownicy_arr = np.array(pracownicy)
max_premia = premie.max()
maska_najlepszych = (premie == max_premia)
najlepsi = pracownicy_arr[maska_najlepszych]

print("\nOsoby z najwyższą premią:")
print(najlepsi)
print("Kwota najwyższej premii:", max_premia)
```

Zadanie 3c: Ranking (argsort)

```python
kolejnosc = np.argsort(pensje)[::-1]


print("Poz.", "Imię", "Pensja", "Staż", sep=" \t ")
print("-" * 40)


for i, idx in enumerate(kolejnosc, 1):
    imie = pracownicy[idx]
    pensja = pensje[idx]
    lata = staz[idx]
    
   
    print(i, imie, pensja, lata, sep=" \t ")
```

Zadanie 3d: Korelacja

```python
kor_staz_pensja = np.corrcoef(staz, pensje)[0, 1]
kor_staz_ocena = np.corrcoef(staz, oceny)[0, 1]


print("Korelacja staż vs pensja:", kor_staz_pensja)
print("Korelacja staż vs ocena:", kor_staz_ocena)
print("-" * 30)


if np.abs(kor_staz_pensja) > np.abs(kor_staz_ocena):
    print("Silniejszy związek występuje między stażem a pensją.")
else:
    print("Silniejszy związek występuje między stażem a oceną.")
```

Zadanie 3e: Analiza sprzedaży (broadcasting)

```python
ceny_kol = ceny_produktow.reshape(4, 1)
przychod = sprzedaz * ceny_kol

suma_produkty = przychod.sum(axis=1)
idx_najlepszy_produkt = np.argmax(suma_produkty)
najlepszy_produkt = produkty_nazwy[idx_najlepszy_produkt]


suma_miesiace = przychod.sum(axis=0)
najlepszy_miesiac_numer = np.argmax(suma_miesiace) + 1  


poczatek_suma = przychod[:, :3].sum()
koniec_suma = przychod[:, 3:].sum()

print("Najlepszy produkt:", najlepszy_produkt)
print("Najlepszy miesiąc (numer):", najlepszy_miesiac_numer)
print("-" * 30)
print("Przychód I połowa:", poczatek_suma)
print("Przychód II połowa:", koniec_suma)

if koniec_suma > poczatek_suma:
    print("Trend jest wzrostowy.")
else:
    print("Trend jest spadkowy.")
```

# Ćwiczenie 4: Generowanie danych

```python
np.random.seed(42)


wiek_raw = np.random.normal(35, 10, 100)

wiek = np.clip(wiek_raw.astype(int), 18, None)


wydatki_raw = np.random.normal(500, 200, 100)

wydatki = np.clip(wydatki_raw, 0, None)


wizyty = np.random.poisson(5, 100)


print("Pierwszych 5 klientów (Wiek | Wydatki | Wizyty):")
print("-" * 45)
for i in range(5):
    print("Klient", i+1, ":", wiek[i], "lat,", wydatki[i], "PLN,", wizyty[i], "wizyt")
```

```python

print("--- STATYSTYKI: WIEK ---")
print("Średnia:", np.mean(wiek).round(1))
print("Mediana:", np.median(wiek))
print("Odchylenie (std):", np.std(wiek).round(1))
print("Min:", np.min(wiek), "| Max:", np.max(wiek))
print()

kor_wiek_wydatki = np.corrcoef(wiek, wydatki)[0, 1]
kor_wizyty_wydatki = np.corrcoef(wizyty, wydatki)[0, 1]

print("Korelacja Wiek vs Wydatki:", kor_wiek_wydatki.round(3))
print("Korelacja Wizyty vs Wydatki:", kor_wizyty_wydatki.round(3))
print()

sr = wydatki.mean()
sd = wydatki.std()

segmenty = np.where(wydatki > sr + sd, "WYSOKI", 
           np.where(wydatki < sr - sd, "NISKI", "ŚREDNI"))

nazwy_segmentow, ilosc = np.unique(segmenty, return_counts=True)

print("--- SEGMENTACJA WYDATKÓW ---")

for i in range(len(nazwy_segmentow)):
    print("Segment:", nazwy_segmentow[i], "| Liczba klientów:", ilosc[i])
```
