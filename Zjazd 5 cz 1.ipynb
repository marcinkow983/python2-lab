Zadanie 1a: Utwórz Series pensje z dict:

```python
dane_pensje = {
    'Anna': 5500,
    'Jan': 7200,
    'Ewa': 4800,
    'Marek': 9100,
    'Kasia': 6300
}

pensje = pd.Series(dane_pensje)
print(pensje)
```

Zadanie 1b: Odpowiedz na pytania o pensje:

```python
najbogatsza_osoba = pensje.idxmax()
print(f"Najwięcej zarabia: {najbogatsza_osoba}")

srednia = pensje.mean()
print(f"Średnia pensja wynosi: {srednia}")

powyzej_sredniej = pensje[pensje > srednia]
print("Osoby zarabiające powyżej średniej:")
print(powyzej_sredniej.index.tolist()) 

liczba_osob = (pensje < 6000).sum()
print(f"Liczba osób zarabiających poniżej 6000: {liczba_osob}")
```

# Krok 2 — Operacje na Series
Zadanie 1c: Oblicz pensje po zmianach:

```python
pensje_podwyzka = pensje * 1.10
print("Pensje po 10% podwyżce:")
print(pensje_podwyzka)

pensje_premia = pensje + 500
print("\nPensje po dodaniu 500 zł premii:")
print(pensje_premia)

nowa_pensja = (pensje * 1.10) + 500
print("\nNowa pensja (podwyżka + premia):")
print(nowa_pensja)

ranking = nowa_pensja.sort_values(ascending=False)
print("\nRanking nowych pensji (od najwyższej):")
print(ranking)
```


Zadanie 3a: Pierwszy rzut oka

```python

print("Pierwsze 10 wierszy:")
print(penguins.head(10)) 

print("\nWymiary tabeli:")
print(penguins.shape)     

print("\nSzczegółowe informacje o zbiorze:")
penguins.info()
```

Zadanie 3b: Statystyki opisowe

```python
print("Podstawowe statystyki (tylko numeryczne):")
display(penguins.describe())

print("\nStatystyki dla wszystkich kolumn (w tym tekstowych):")
display(penguins.describe(include='all'))

mediana_masy = penguins['body_mass_g'].median()
print(f"\nMediana masy ciała wynosi: {mediana_masy} g")
```

Zadanie 3c: Rozkłady wartości

```python

print("Liczba pingwinów według gatunku:")
print(penguins['species'].value_counts())

print("\nLiczba pingwinów na każdej wyspie:")
print(penguins['island'].value_counts())

print("\nRozkład płci:")
print(penguins['sex'].value_counts())

print("\nProcentowy udział gatunków:")
procenty = penguins['species'].value_counts(normalize=True) * 100
print(procenty)
```

Zadanie 3d: Brakujące dane

```python

braki_suma = penguins.isna().sum()
print("Liczba brakujących wartości w poszczególnych kolumnach:")
print(braki_suma)

braki_sex = penguins['sex'].isna().sum()
procent_brakow = (braki_sex / len(penguins)) * 100

print(f"\nW kolumnie 'sex' brakuje {procent_brakow:.2f}% danych.")
```
