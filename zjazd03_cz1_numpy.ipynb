# Operacje wektorowe — analiza biznesowa

Produkty w sklepie internetowym

```python
nazwy = ['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka',
         'Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka']

ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

Zadanie 1: Podstawowe statystyki cenowe

```python
srednia_wartosc = ceny.mean()
indeks_max = ceny.argmax()
indeks_min = ceny.argmin()

print("Średnia cena produktów:", srednia_wartosc)
print("Najdroższy produkt:", nazwy[indeks_max], "w cenie", ceny[indeks_max])
print("Najtańszy produkt:", nazwy[indeks_min], "w cenie", ceny[indeks_min])
```

Zadanie 2: Wartość magazynu

```python
wartosci_produktow = ceny * ilosci_magazyn


laczna_wartosc_magazynu = wartosci_produktow.sum()

print("Wartości poszczególnych produktów w magazynie:")
print(wartosci_produktow)

print("\nŁączna wartość całego magazynu wynosi:", laczna_wartosc_magazynu, "zł")
```

Zadanie 3: Analiza cenowa

```python
cena_po_rabacie= ceny * 0.85
print("cena po rabacie wynosi: ",cena_po_rabacie)

cena_Vat = ceny * 1.23
print("cena z VAT wynosi: ",cena_Vat)

ile_powyzej = (ceny > 1000).sum()
print("liczba produktów powyżej 1000zł: ",ile_powyzej)
```

Zadanie 4: Filtrowanie

```python
nazwy_np = np.array(nazwy)


oceny = oceny_klientow >= 4.5
print("Produkty z oceną >= 4.5:")
print("Nazwy:", nazwy_np[oceny])
print("Ceny:", ceny[oceny])


cena = ceny < 200
print("\nProdukty poniżej 200 zł:")
print("Nazwy:", nazwy_np[cena])
print("Ilości:", ilosci_magazyn[cena])
```

Zadanie 5: Analiza zapasów

```python
nazwy_np = np.array(nazwy)
wartosci_wszystkie = ceny * ilosci_magazyn


ilosc_niski_stan = ilosci_magazyn < 30
nazwy_niski_stan = nazwy_np[ilosc_niski_stan]

wartosc_niski_stan = wartosci_wszystkie[ilosc_niski_stan].sum()

laczna_wartosc = wartosci_wszystkie.sum()
procent_wartosci = (wartosc_niski_stan / laczna_wartosc) * 100

print("Produkty z niskim stanem magazynowym:")
print(nazwy_niski_stan)

print("\nŁączna wartość tych produktów:", wartosc_niski_stan, "zł")
print("Stanowią one", procent_wartosci, "% wartości całego magazynu")
```

Zadanie 6 (bonus): Analiza korelacji cena–ocena

```python
nazwy_np = np.array(nazwy)


okazja = (oceny_klientow >= 4.5) & (ceny < 1000)

print("Produkty będące okazją:")
print("Nazwy:", nazwy_np[okazja])
print("Ceny:", ceny[okazja])

wysoka_ocena = oceny_klientow >= 4.0
srednia_wysoka = ceny[wysoka_ocena].mean()

niska_ocena = oceny_klientow < 4.0
srednia_niska = ceny[niska_ocena].mean()

print("\nPorównanie średnich cen:")
print("Średnia cena produktów z oceną >= 4.0:", srednia_wysoka, "zł")
print("Średnia cena produktów z oceną < 4.0:", srednia_niska, "zł")
```

# Benchmark — lista vs NumPy

```python
import time

# Przygotowanie danych
rozmiar = 1_000_000
lista = list(range(rozmiar))
tablica = np.array(lista)

# Test 1: Pomnóż przez 2
start = time.perf_counter()
wynik_lista = [x * 2 for x in lista]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
wynik_numpy = tablica * 2
czas_numpy = time.perf_counter() - start

print(f"=== Mnożenie ×2 ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
# Test 2: Suma
start = time.perf_counter()
suma_lista = sum(lista)
czas_lista = time.perf_counter() - start

start = time.perf_counter()
suma_numpy = tablica.sum()
czas_numpy = time.perf_counter() - start

print(f"\n=== Suma ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
# Test 3: Filtrowanie (elementy > 500000)
start = time.perf_counter()
filtr_lista = [x for x in lista if x > 500000]
czas_lista = time.perf_counter() - start

start = time.perf_counter()
filtr_numpy = tablica[tablica > 500000]
czas_numpy = time.perf_counter() - start

print(f"\n=== Filtrowanie > 500k ({rozmiar:,} elementów) ===")
print(f"Lista:  {czas_lista*1000:.1f} ms")
print(f"NumPy:  {czas_numpy*1000:.1f} ms")
print(f"NumPy jest {czas_lista/czas_numpy:.0f}× szybszy!")
```
